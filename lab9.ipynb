```python
#q4

student_data = {
    "Name": [
        "Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hank", "Ivy", "Jack",
        "Kara", "Leo", "Mia", "Nina", "Oscar", "Paul", "Quinn", "Rose", "Sam", "Tina"
    ],
    "Gender": [
        "Male", "Female", "Male", "Male", "Male", "Female", "Male", "Male", "Male", "Female",
        "Male", "Male", "Male", "Male", "Female", "Male", "Female", "Female", "Female", "Male"
    ],
    "Age": [19, 26, 33, 45, 20, 28, 13, 33, 36, 22, 55, 42, 18, 25, 35, 32, 22, 44, 54, 53],
    "Height (cm)": [
        141.9, 179.0, 149.3, 143.6, 164.7, 155.3, 179.3, 158.7, 174.4, 167.2,
        158.0, 140.5, 177.7, 162.5, 155.4, 140.6, 149.2, 149.6, 167.3, 164.4
    ],
    "Math Marks": [47, 86, 74, 53, 56, 75, 89, 79, 43, 41, 45, 93, 81, 43, 93, 68, 57, 65, 83, 73],
    "Science Marks": [49, 75, 53, 70, 87, 54, 47, 53, 62, 99, 96, 79, 60, 55, 84, 57, 86, 92, 63, 65]
}

df_students = pd.DataFrame(student_data)

df_students['Age Group'] = ""

for index, age in enumerate(df_students["Age"]):
    if age >= 18 and age <= 25:
        df_students.at[index, 'Age Group'] = "18-25"
    elif age >= 26 and age <= 30:
        df_students.at[index, 'Age Group'] = "26-30"
    elif age >= 31 and age <= 35:
        df_students.at[index, 'Age Group'] = "31-35"
    elif age >= 36:
        df_students.at[index, 'Age Group'] = "36 and above"

plt.pie(df_students["Age Group"].value_counts(),labels=df_students["Age Group"].value_counts().index)
plt.show()
```

```python
#q5
plt.pie(df_students["Gender"].value_counts(),labels=df_students["Gender"].value_counts().index,autopct="%1.1f%%")
```

```python
#q6
plt.scatter(df_students["Math Marks"],df_students["Science Marks"])
plt.xlabel("maths")
plt.ylabel("science")
plt.legend(["Students"], loc="upper left")
plt.show()
```
